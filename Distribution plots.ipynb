```python
import seaborn as sns
```

```python
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
```

### distribution plot

```python
sns.distplot(tips['tip'],kde=False,color = 'r', bins = 10)
```

### Joint plot

```python
tips.head()
```

```python
##plotting tips data inot joint plot
sns.jointplot(x = 'total_bill',y='tip', data = tips, kind = 'scatter')

# arguments recall
# kind = 'hex' for hexagon , 'reg' -> for regression line , kde = 2d kde 
```

# Pair plot

```python
sns.pairplot(tips, hue = 'sex')
```

### rugplot()

```python
sns.rugplot(tips.total_bill)
```
