```python
import seaborn as sns 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc, annot = True, cmap ='Purples')  #you can pass in any colormap (cmap) eg : Reds ,Purples, coolwarm , Blues ,.........many more
plt.show()
```

```python
m = flights.corr()

sns.heatmap(m, annot = True)
```

```python
## --------------------------- pivot_table -----------------------------

# In pandas, the pivot_table() function is available to create pivot tables. It takes a DataFrame as input and allows you 
#to specify the index, columns, and values to aggregate. 

fp = flights.pivot_table(index = 'month', columns = 'year', values ='passengers')
fp
```

```python
sns.heatmap(fp, cmap = 'coolwarm', linecolor = 'white', linewidth = .5)

plt.show()
```

# clustermap 

```python
sns.clustermap(fp, standard_scale = 1)
```
