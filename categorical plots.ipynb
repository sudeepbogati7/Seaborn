# Different categorical plots :

##### 1.Bar plot
##### 2.Count plot
##### 3. Box plot
##### 4. Violin plot
##### 5. Strip plot 
##### 6. Swarm plot
##### 7.Catplot --(factor plot in past)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
planets = sns.load_dataset('planets')


tips = sns.load_dataset('tips')
```

### 1. Bar plot

```python
sns.barplot(x= 'sex', y='total_bill',data = tips, estimator = np.std )
```

### 2. Count plot

```python
sns.countplot(x='sex', data =tips)
```

## 3. Box plots

```python
sns.boxplot(x='day',y='total_bill' , data = tips, hue ='smoker')
```

### 4. Violin plot

```python
sns.violinplot(x='day', y='total_bill',data=tips,hue ='sex', split= True)
```

## Comparision between box and violin plot

```python
fig = plt.figure(figsize = (10,6))

fig, axes = plt.subplots(nrows =1 , ncols =2)
sns.boxplot(x='day', y='total_bill',data=tips,hue ='sex',ax =axes[1] )
sns.violinplot(x='day', y='total_bill', data=tips,hue='sex', ax =axes[0])

plt.tight_layout()
```

## 5. strip plots

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue ='smoker', dodge =True)
plt.show()
```

## 6. swarm plot   
###### similar to strip plot
###### --> inconvinient for larger data sets though

```python
sns.swarmplot(x= 'day',y='total_bill', data = tips,hue ='sex', dodge = True)  ## split is renamed as dodge
```

##### We can combine swarm plot and violin plot as : 


```python
sns.swarmplot(x= 'day',y='total_bill', data = tips, color = 'black')
sns.violinplot(x= 'day',y='total_bill', data = tips)
```

## 7. Factor plot:
##### In now 2023, factor plot has been renamed as catplot()


```python
sns.catplot(x= 'day',y='total_bill', data = tips, kind='violin') ## you can specify any kind = '' arguments , which plot you want to create such as 'bar', 'violin'
```
