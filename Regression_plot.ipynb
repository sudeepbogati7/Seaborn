```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# lmplot(x = 'x-axis', y ='y-axis', data ='source data')

sns.lmplot(x ='total_bill', y ='tip', data =tips,hue ='sex' , markers =['v','o'], col = 'sex', row ='time')
```

```python
sns.lmplot(x ='total_bill', y ='tip', data =tips , markers =['v','o'], col = 'day', hue ='sex',
           aspect = 0.6)
```
